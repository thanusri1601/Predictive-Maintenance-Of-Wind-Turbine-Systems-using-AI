```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, LSTM, Dense, Dropout, Input
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 🟢 Step 1: Load Data Efficiently
def reduce_memory_usage(df):
    """Reduce memory usage of a DataFrame by optimizing dtypes."""
    for col in df.columns:
        if df[col].dtype == "float64":
            df[col] = df[col].astype("float32")
        elif df[col].dtype == "int64":
            df[col] = df[col].astype("int32")
    return df

# Assuming `data` is your DataFrame
data = reduce_memory_usage(data)

# 🟢 Step 2: Define Target (y) and Features (X)
target_col = 'Total reactive power avg'
drop_cols = [
    'asset_id', 'event_description', target_col, 'id', 'train_test', 'status_type_id',
    'event_id', 'event_label', 'time_stamp',
    'Possible Grid capacitive reactive power avg',
    'Possible Grid capacitive reactive power max',
    'Possible Grid capacitive reactive power min',
    'Possible Grid capacitive reactive power std',
    'Possible Grid inductive reactive power avg',
    'Possible Grid inductive reactive power max',
    'Possible Grid inductive reactive power min',
    'Possible Grid inductive reactive power std',
    'Possible Grid active power avg', 'Possible Grid active power max',
    'Possible Grid active power min', 'Possible Grid active power std',
    'Grid power avg', 'Grid power max', 'Grid power min', 'Grid power std',
    'Grid reactive power avg', 'Grid reactive power max',
    'Grid reactive power min', 'Grid reactive power std',
    'Generator disconnected active power avg',
    'Generator connected in delta active power avg',
    'Generator connected in star active power avg',
    'Generator disconnected reactive power avg',
    'Generator connected in delta reactive power avg',
    'Generator connected in star reactive power avg',
    'Total active power avg', 'Total reactive power avg',
]

X = data.drop(columns=drop_cols, errors='ignore')
y = data[target_col].values

# 🟢 Step 3: Normalize and Reshape Data Efficiently
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reduce dataset size using downsampling (Optional)
num_samples = 500000  # Adjust based on RAM availability
X_scaled = X_scaled[:num_samples]
y = y[:num_samples]

# Reshape for CNN-LSTM (batch_size, time_steps, features)
time_steps = 10
X_seq = np.array([X_scaled[i-time_steps:i] for i in range(time_steps, len(X_scaled))])
y_seq = y[time_steps:]  # Align target variable

# 🟢 Step 4: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_seq, y_seq, test_size=0.2, random_state=42)

# 🟢 Step 5: Define CNN-LSTM Model (Fixed Input Shape)
input_layer = Input(shape=(time_steps, X_train.shape[2]))

x = Conv1D(filters=64, kernel_size=3, activation='relu')(input_layer)
x = MaxPooling1D(pool_size=2)(x)
x = LSTM(50, return_sequences=True)(x)
x = LSTM(50)(x)
x = Dense(50, activation='relu')(x)
x = Dropout(0.3)(x)
output_layer = Dense(1)(x)

model = Model(inputs=input_layer, outputs=output_layer)
model.compile(optimizer='adam', loss='mse')

# Train model
model.fit(X_train, y_train, epochs=2, batch_size=512, validation_data=(X_test, y_test))

# 🟢 Step 6: Ensure Model is Initialized
dummy_input = np.zeros((1, X_train.shape[1], X_train.shape[2]))
_ = model.predict(dummy_input)

# 🟢 Step 7: Extract Features from CNN-LSTM
feature_extractor = Model(inputs=model.input, outputs=model.layers[-2].output)
X_train_features = feature_extractor.predict(X_train, batch_size=512)
X_test_features = feature_extractor.predict(X_test, batch_size=512)

# 🟢 Step 8: Train XGBoost Model
xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=6, tree_method='hist')

xgb_model.fit(X_train_features, y_train)

# 🟢 Step 9: Predictions & Evaluation
y_pred = xgb_model.predict(X_test_features)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f'MAE: {mae:.4f}')
print(f'RMSE: {rmse:.4f}')
print(f'R² Score: {r2:.4f}')

# 🟢 Step 10: Visualization
plt.figure(figsize=(10, 6))
plt.plot(y_test[:500], label='Actual Reactive Power', linestyle='dashed')
plt.plot(y_pred[:500], label='Predicted Reactive Power')
plt.title('Actual vs Predicted Reactive Power')
plt.xlabel('Samples')
plt.ylabel('Reactive Power')
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from keras.models import Sequential
from keras import layers
from keras.callbacks import EarlyStopping

# Function to create sequences
def create_sequences(data, target, time_steps):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:i + time_steps, :])
        y.append(target[i + time_steps - 1])
    return np.array(X), np.array(y)
target_col = 'Total reactive power avg'
drop_cols = [
    'asset_id', 'event_description', target_col, 'id', 'train_test', 'status_type_id',
    'event_id', 'event_label', 'time_stamp',
    'Possible Grid capacitive reactive power avg',
    'Possible Grid capacitive reactive power max',
    'Possible Grid capacitive reactive power min',
    'Possible Grid capacitive reactive power std',
    'Possible Grid inductive reactive power avg',
    'Possible Grid inductive reactive power max',
    'Possible Grid inductive reactive power min',
    'Possible Grid inductive reactive power std',
    'Possible Grid active power avg', 'Possible Grid active power max',
    'Possible Grid active power min', 'Possible Grid active power std',
    'Grid power avg', 'Grid power max', 'Grid power min', 'Grid power std',
    'Grid reactive power avg', 'Grid reactive power max',
    'Grid reactive power min', 'Grid reactive power std',
    'Generator disconnected active power avg',
    'Generator connected in delta active power avg',
    'Generator connected in star active power avg',
    'Generator disconnected reactive power avg',
    'Generator connected in delta reactive power avg',
    'Generator connected in star reactive power avg',
    'Total active power avg', 'Total reactive power avg',
]

X = data.drop(columns=drop_cols, errors='ignore')
y = input_data['RUL'].values

# Data Preprocessing
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create sequences
time_steps = 15  # Increased time_steps for better temporal learning
X_seq, y_seq = create_sequences(X_scaled, y, time_steps)
y_seq = y_seq.reshape(-1, 1)

# CNN-LSTM Model
cnn_lstm_model = Sequential([
    layers.Input(shape=(time_steps, X_seq.shape[2])),

    layers.Conv1D(64, kernel_size=3, activation='relu'),  # Increased filters
    layers.Conv1D(32, kernel_size=3, activation='relu'),
    layers.MaxPooling1D(pool_size=2),

    layers.Bidirectional(layers.LSTM(128, return_sequences=True)),  # Added Bidirectional LSTM
    layers.LSTM(64, return_sequences=False),

    layers.Dropout(0.4),  # Increased dropout to prevent overfitting

    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)  # Single output neuron
])

cnn_lstm_model.compile(optimizer='adam', loss='huber', metrics=['mae'])  # Changed loss function

# Early stopping
early_stopping = EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)

# Train model
cnn_lstm_model.fit(X_seq, y_seq, epochs=50, batch_size=64, callbacks=[early_stopping], verbose=1)  # Increased epochs

# Extract learned features
X_cnn_lstm_features = cnn_lstm_model.predict(X_seq)

# XGBoost Model
xgb_model = xgb.XGBRegressor(
    objective='reg:squarederror',
    eval_metric='mae',
    n_estimators=500,  # Increased estimators
    learning_rate=0.05,  # Lowered learning rate
    max_depth=6,  # Added depth for better decision making
    subsample=0.8,  # Added subsampling
    colsample_bytree=0.8
)

xgb_model.fit(X_cnn_lstm_features, y_seq, eval_set=[(X_cnn_lstm_features, y_seq)], verbose=True)

# Predict
y_pred = xgb_model.predict(X_cnn_lstm_features)

# Evaluate
mae = mean_absolute_error(y_seq, y_pred)
rmse = np.sqrt(mean_squared_error(y_seq, y_pred))
r2 = r2_score(y_seq, y_pred)

print(f'MAE: {mae}')
print(f'RMSE: {rmse}')
print(f'R² Score: {r2}')

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(y_seq, label='Actual RUL')
plt.plot(y_pred, label='Predicted RUL')
plt.title('Actual vs Predicted RUL')
plt.xlabel('Samples')
plt.ylabel('RUL')
plt.legend()
plt.show()
```
