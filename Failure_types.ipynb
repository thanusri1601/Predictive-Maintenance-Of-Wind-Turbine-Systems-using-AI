```python
input_data = data.drop(['time_stamp'], axis=1)
```

```python
#Handling missing values
input_data=input_data.dropna()
data=data.dropna()
missing_data=data.isnull().sum()
missing_values=input_data.isnull().sum()
print("Missing values: \n", missing_data[missing_data>0])
print("Missing values: \n", missing_values[missing_values>0])
```

```python
input_data['event_description'] = input_data['event_description'].str.strip().map({
    'Normal': 0,
    'Generator bearing failure': 1,
    'Hydraulic group': 2,
    'Transformer failure': 3,
    'Gearbox failure': 4
})
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from xgboost import plot_importance

# Load your dataset (assuming it's already in 'input_data')
df = input_data.copy()

# Loop through each failure mode (1, 2, 3, 4)
for failure_mode in [1, 2, 3, 4]:
    print(f"\nFeature Importance for Event Description = {failure_mode}\n")

    # Filter dataset (include only rows where event_description is 0 or the specific failure mode)
    df_filtered = df[df['event_description'].isin([0, failure_mode])]

    # Define Features (X) and Target (y)
    X = df_filtered.drop(columns=['event_description', 'RUL', 'status_type_id', 'asset_id'], errors='ignore')  # Drop RUL if present
    y = (df_filtered['event_description'] == failure_mode).astype(int)  # Convert to binary (0 = normal, 1 = failure mode)

    # Train XGBoost Classifier
    model = xgb.XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=6, random_state=42)
    model.fit(X, y)

    # Get feature importance
    feature_importance = pd.Series(model.feature_importances_, index=X.columns)

    # Get top 20 important features
    top_features = feature_importance.nlargest(20)

    # Print top features
    print(top_features)

    # Plot feature importance
    plt.figure(figsize=(10, 6))
    top_features.sort_values().plot(kind='barh', color='steelblue')
    plt.xlabel("Feature Importance Score")
    plt.ylabel("Sensor Features")
    plt.title(f"Top 20 Features for Event Description = {failure_mode}")
    plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier

# Load your dataset (assuming it's already in 'input_data')
df = input_data.copy()

# Loop through each failure mode (1 to 4)
for failure_mode in [1, 2, 3, 4]:
    print(f"\nFeature Importance for Event Description = {failure_mode}\n")

    # Filter dataset (include only rows where event_description is 0 or the specific failure mode)
    df_filtered = df[df['event_description'].isin([0, failure_mode])]

    # Define Features (X) and Target (y)
    X = df_filtered.drop(columns=['event_description', 'RUL', 'status_type_id', 'asset_id'])  # Sensor values
    y = df_filtered['event_description']  # Target (Normal = 0, Failure Mode = 1, 2, 3, or 4)

    # Train Random Forest Classifier
    clf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf.fit(X, y)

    # Get feature importance
    feature_importance = pd.Series(clf.feature_importances_, index=X.columns)

    # Get top 20 important features
    top_features = feature_importance.nlargest(20)

    # Print top features
    print(top_features)

    # Plot feature importance
    plt.figure(figsize=(10, 6))
    top_features.sort_values().plot(kind='barh', color='steelblue')
    plt.xlabel("Feature Importance Score")
    plt.ylabel("Sensor Features")
    plt.title(f"Top 20 Features for Event Description = {failure_mode}")
    plt.show()
```
